```python
import numpy as np
import pandas as pd
import datetime
```

```python
# Read the CSV file
data = pd.read_csv('bank_marketing.csv')
data
```

```python
# Check if there are duplicated values in the data set
data.duplicated()
```

```python
# Remove the duplicates in the dataset which targets the primary key client_id
data = data.drop_duplicates(subset=['client_id'], keep='first').copy()
data
```

```python
# Define a mapping of month names to numerical values
month_mapping = {
    'jan': 1, 'feb': 2, 'mar': 3, 'apr': 4, 'may': 5, 'jun': 6,
    'jul': 7, 'aug': 8, 'sep': 9, 'oct': 10, 'nov': 11, 'dec': 12
}
```

```python
# Add a temporary year column (2022) that will be used for formatting later
data['year'] = 2022
```

```python
# Replace month names with numeric values
data['month'] = data['month'].map(month_mapping)
data
```

```python
# Combine year, month, and day into a single date column
data['date'] = pd.to_datetime(data[['year', 'month', 'day']])
data
```

```python
# Drop unnecessary and temporary columns
data = data.drop(columns=['month', 'year', 'day'])
data
```

```python
# Convert the values of 'contact_duration' into numeric 
data['contact_duration'] = pd.to_numeric(data['contact_duration'], errors='coerce')
data
```

```python
# Convert seconds to 'MM:SS' format and overwrite the original column
data['contact_duration'] = data['contact_duration'].apply(
    lambda x: str(datetime.timedelta(seconds=int(x)))[-5:] if not pd.isnull(x) else '00:00'
)
data
```

```python
# Convert 'campaign_outcome' values from 'yes'/'no' to 'Successful'/'Failed'
data['campaign_outcome'] = data['campaign_outcome'].map({'yes': 'Successful', 'no': 'Failed'})
data
```

```python
# Capitalize first letter of every word in all string columns
data = data.apply(lambda x: x.str.title() if x.dtype == "object" else x)
data
```

```python
# Replace "Unknown" values with NaN
data = data.replace("Unknown", np.nan)
data
```
